```python
# Aldar Properties ‚Äì Sales & Market Insight  
# Tools: Python, pandas, matplotlib

import pandas as pd
import matplotlib.pyplot as plt

# 1Ô∏è‚É£ Published data (manually entered)  
data = {
    "Period": ["FY 2023", "FY 2024", "Q1 2025"],
    "Development_Sales_AED_bn": [28.0, 33.6, 8.9],  # e.g. 2023 approx, 2024 actual, Q12025
    "YoY_Change_%": [None, 20.0, 42.0]  # 2024 vs 2023 = +20%; Q12025 vs Q12024 = +42%
}

df = pd.DataFrame(data)
print("Dataset preview:")
display(df.fillna(0))

# 2Ô∏è‚É£ Visual: Sales growth by period  
plt.figure(figsize=(6,4))
plt.plot(df["Period"], df["Development_Sales_AED_bn"], marker="o", linestyle="-", color="darkblue")
for i,v in enumerate(df["Development_Sales_AED_bn"]):
    plt.text(i, v+0.5, f"{v:.1f}", ha="center")
plt.title("Aldar Properties ‚Äì Development Sales (AED bn)")
plt.ylabel("Sales (AED bn)")
plt.xlabel("Period")
plt.tight_layout()
plt.savefig("aldar_sales_trend.png", dpi=150)
plt.show()

# 3Ô∏è‚É£ Insights
print("\nüîç Key Insights:")
print(f"- In FY2024, Aldar achieved AED {df.loc[1,'Development_Sales_AED_bn']:.1f} bn in development sales, up ~{df.loc[1,'YoY_Change_%']:.0f}% year-on-year.")
print(f"- In Q1 2025 alone, sales reached AED {df.loc[2,'Development_Sales_AED_bn']:.1f} bn (a ~{df.loc[2,'YoY_Change_%']:.0f}% rise vs Q1 2024).")
print("- Strong upward momentum suggests robust demand, especially international buyers and off-plan launches in Abu Dhabi.")

# 4Ô∏è‚É£ Recommendation (markdown style)
# You can add this as a markdown cell in your notebook:
#
# **Recommendation:**  
# With this accelerated growth, Aldar should continue to:
# - Expand its international marketing for high-net-worth investors  
# - Launch more premium residential projects with ‚Äúfast-sell‚Äù timelines  
# - Maintain a focus on Abu Dhabi and leverage favourable regulation to sustain momentum  
```
